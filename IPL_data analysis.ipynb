## 9.In which years where taking batting second have won more number of matches?

```python
field_wins = ipl[ipl['toss_decision'] == 'field']['winner'].count()

# Count the number of matches won by teams choosing to bat first
bat_wins = ipl[ipl['toss_decision'] == 'bat']['winner'].count()

# Compare the counts
if field_wins > bat_wins:
    print("Teams winning tosses and electing field first have won more number of matches.")
elif bat_wins > field_wins:
    print("Teams winning tosses and electing bat first have won more number of matches.")
else:
    print("Both A and B (Teams winning tosses and electing field first or bat first) have won an equal number ofmatches")
```

```python
batting_second_wins = ipl[(ipl['inning'] == 2) | (ipl['toss_decision'] == 'field')]

# Count the number of matches won by teams batting second for each year
batting_second_wins_by_year = batting_second_wins.groupby('season')['winner'].count()

# Identify the years where teams batting second won more matches
years_with_more_batting_second_wins = batting_second_wins_by_year[batting_second_wins_by_year.idxmax()]

print("Years where teams batting second won more matches:")
print(years_with_more_batting_second_wins)
```

```python
years_with_more_batting_second_wins = batting_second_wins_by_year[batting_second_wins_by_year == batting_second_wins_by_year.max()]

print("Years where teams batting second won more matches:")
print(years_with_more_batting_second_wins)
```

## 11. In 2019 which batsman scored highest number of runs by hitting 6's and 4's?

```python
matches_2019['fours'] = matches_2019['batsman_runs'].apply(lambda x: 1 if x == 4 else 0)
matches_2019['sixes'] = matches_2019['batsman_runs'].apply(lambda x: 1 if x == 6 else 0)
matches_2019['total_runs_boundary'] = matches_2019['fours'] * 4 + matches_2019['sixes'] * 6

# Find the batsman who scored the highest number of runs by hitting 6's and 4's in 2019
top_batsman_2019 = matches_2019.groupby('batsman')['total_runs_boundary'].sum().idxmax()

print(f"In 2019, the batsman who scored the highest number of runs by hitting 6's and 4's: {top_batsman_2019}")
```

## 13.  What is the strike rate of Kohli in 2016


```python
kohli_2016 = ipl[(ipl['batsman'] == 'V Kohli') & (ipl['season'] == 2016)]
kohli_2016
```

```python
# Calculate the total runs and total balls faced
total_runs = kohli_2016['total_runs'].sum()
total_balls = kohli_2016['ball'].sum()

# Calculate the strike rate
strike_rate = (total_runs / total_balls) * 100

print(f"Virat Kohli's strike rate in 2016: {strike_rate:.2f}")
```

## 14.  Bowlers with maximum number of extras:

## 19. Across all seasons which player was dismissed the maximum number of times via caught and bowled


```python
# Filter the DataFrame for dismissals via caught and bowled
caught_and_bowled_dismissals = ipl[ipl['dismissal_kind'] == 'caught and bowled']

# Group by 'batsman' and count the number of caught and bowled dismissals
caught_and_bowled_counts = caught_and_bowled_dismissals['batsman'].value_counts()

# Get the player with the maximum number of caught and bowled dismissals
max_caught_and_bowled_player = caught_and_bowled_counts.idxmax()
max_caught_and_bowled_count = caught_and_bowled_counts.max()

print(f"The player dismissed the maximum number of times via caught and bowled is: {max_caught_and_bowled_player}")
print(f"The number of times: {max_caught_and_bowled_count}")
```
