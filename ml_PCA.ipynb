```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
df.head(10)
```

```python
df.shape
```

```python
z_scores = np.abs((df['Cholesterol'] - df['Cholesterol'].mean()) / df['Cholesterol'].std())
df_filtered = df[z_scores <= 3]
```

```python
df_filtered.shape
```

```python
df_filtered['Sex'] = df_filtered['Sex'].apply(lambda x: 1 if x == 'M' else 0)
df_filtered['ExerciseAngina'] = df_filtered['ExerciseAngina'].apply(lambda x: 1 if x == 'Y' else 0)
df_filtered['ST_Slope'] = df_filtered['ST_Slope'].apply(lambda x: 1 if x == 'Up' else 0)

df_filtered
```

```python
df1 = pd.get_dummies(df_filtered,drop_first=True)
df1
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Assuming you have your feature matrix X and target variable y

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df1.drop('HeartDisease',axis=1), df1.HeartDisease, test_size=0.2, random_state=42)

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
model.predict(X_test)

# Evaluate the model's accuracy
model.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Assuming you have your feature matrix X and target variable y

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df1.drop('HeartDisease',axis=1), df1.HeartDisease, test_size=0.2, random_state=42)

# Create an SVM classifier
model = SVC(kernel='rbf', C=1.0, gamma='scale')

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming you have your feature matrix X

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df1.drop('HeartDisease',axis=1))

# Create a PCA object
pca = PCA(0.95)  # Specify the number of components to keep

# Perform PCA on the standardized data
X_pca = pca.fit_transform(X_scaled)

# Access the principal components
principal_components = pca.components_

# Access the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Print the explained variance ratio
print("Explained variance ratio:", explained_variance_ratio)

# Access the transformed data
print("Transformed data shape:", X_pca.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Assuming you have your feature matrix X and target variable y

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, df1.HeartDisease, test_size=0.2, random_state=42)

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
model.predict(X_test)

# Evaluate the model's accuracy
model.score(X_test,y_test)
```
